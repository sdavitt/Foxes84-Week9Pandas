### Data Visualization <br>
<p>Data visualization is one of the most powerful tools an analyst has for two main reasons. Firstly, it is unrivalled in it's ability to guide the analysts hand in determining 'what to look at next'. Often a visual is revealing of patterns in the data that are not easily discernable by just looking at dataframes. Secondly, they are an analysts greatest communication tool. Very often professional analysts need to present their results to groups of people responsible for taking action based on what the data says. Visuals can tell your story much better then raw numbers, so let's talk a little bit about how to make effective visuals that make a point.</p>

### Different Types of Charts <br>
<ul>
    <li><b>Line Chart:</b> Exploring data over time</li>
    <li><b>Bar Chart:</b> Comparing categories of data and tracks changes over time</li>
    <li><b>Pie Chart:</b> Explore parts of a whole, i.e. fractions</li>
    <li><b>Scatter Plot:</b> Similar to line charts, track correlations between two categories</li>
    <li><b>Histogram:</b> Unrelated from bar charts, show distribution of variables</li>
    <li><b>Candlestick Chart:</b> Used a lot in financial sector, i.e. can compare a stock over a period of time</li>
    <li><b>Box Chart:</b> Looks identical to candlestick charts, compares minimum, 1st, medium, 3rd quartiles, and max values</li>
</ul>

<p>Depending on what you need to accomplish in conceptualizing your data, you will go with a specific type of chart to portray your data.</p>

### Basics <br>
<p>Plots can be as detailed as you'd like them to be, but the simplest plot to start with can be done in just a few lines:</p>

```python
# initializes a plot object
# then we can modify that plot object
# then we can show or save that plot object
plt.plot([3, 12, 58, 29, 21, 3, 45, 67, 2, 59]) # default create a line graph
# many plt functions simply act on the most recently create plot
# so we just created a plot on line 4
# any next plt functions we run will act on that plot

# call plt.show() to show the most recently created plot
# in this case a basic default line graph
plt.show()
```

### Customization <br>
<p>There's many customizable options when using Matplotlib, such as; line types, colors, labels, figure size, etc. We'll go over a few of them here:</p>

##### Axis Labels & Line Marks

```python
# call to plt.plot() to create a line for the next line graph
# plt.plot([<xvals>], [<yvals>], <linecustomization>)
plt.plot([1, 2, 3, 4, 5], [23, 45, 67, 78, 89], '*-y') # format string for line format '[marker][line][color]'
plt.plot([3, 4, 5, 6, 7, 8], [22, 33, 90, 105, 110, 120], 'x:b')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

##### Legend & Line Labels

```python
plt.plot([1600, 1700, 1800, 1900, 2000], [2, 2, 3, 3.1, 2.4], 'o-r', label='Fox')
plt.plot([1600, 1700, 1800, 1900, 2000], [4, 5, 4, 3.8, 4.5], '^:g', label='Chicken')
plt.ylabel('Population (Millions)')
plt.xlabel('Year')
plt.legend() # adds a legend showing the line labels (position is chosen by default by the function, can be specified)
plt.show()
```

### Subplots <br>
<p>Using the subplots method, you're able to create multiple plots within a single cell.</p>

```python
# setting up a subplot:
# plt.subplot(<total rows>, <total columns>, <this subplot's position>)
# 2 subplots - one for fox one for chicken

# first subplot
p1 = plt.subplot(2, 1, 2)
plt.plot([1600, 1700, 1800, 1900, 2000], [2, 2, 3, 3.1, 2.4], 'o-r', label='Fox')
plt.ylabel('Population (Millions)')
plt.xlabel('Year')
plt.legend()
# modify scale of axes through plt.set_xlim([<min>, <max>]) plt.set_ylim([<min>, <max>])
plt.ylim([1.8, 5.2])

# second subplot
p2 = plt.subplot(2, 1, 1)
plt.plot([1600, 1700, 1800, 1900, 2000], [4, 5, 4, 3.8, 4.5], '^:g', label='Chicken')
plt.legend()
plt.ylim([1.8, 5.2])

# show both plots
plt.show()
```

### Presentation Worthiness
<p>Before we go into how to use this as a tool for data exploration, let's talk quickly about presentation-worthiness. In the next cell we're going to introduce customization of lines, colors, ticks, titles, fonts, and grids.</p>

```python
# define some NDarrays to use with the below customization
x_axis = np.linspace(-10,10,num=20)
y_axis = x_axis**2

fig, ax = plt.subplots(figsize = (20, 10))

plt.plot(x_axis ** 2, y_axis ** 4,
        lw = 5, # Line Width
        ls = '--', # Line Style
        color = 'r', # Line Color
        marker = "*", #Pointer Marker
        markersize = 15, # Marker size
        markerfacecolor = 'b', # Marker Color
        label = 'Data'
        )

plt.xlabel(r'$x$ Value', fontname = 'Times New Roman', fontsize = 32)
plt.ylabel(r'$y$ Value', fontname = 'Times New Roman', fontsize = 32)

plt.xticks(fontsize = 28)
plt.yticks(fontsize = 28)

plt.title('Quadratic Equation Visual', fontname = "Times New Roman", fontsize = 40, y = 1.05)

plt.grid(True)

plt.legend(loc = 'upper left', fontsize = 40)

plt.tight_layout()
```

<p>It's sometimes nice to obsess a little bit over how your plots look. It's good practice paying attention to detail and it has huge payoff when you end up presenting your findings to other analysts and business parteners. Here's a more complex example of using the matplotlib api:</p>

```python
markers = ['o', 'v', '^', '*', 's']
styles = ['-', '--', '-.', ':', '-']
colors = ['#588C7E', '#F2E394', '#F2AE72', '#D96459', '#8C4646']
```

```python
x = np.linspace(0, 10, 50)

fig = plt.subplots(figsize = (20, 10))

for i in range(5):
    plt.plot(x, x ** i,
            lw = 5,
            ls = styles[i],
            color = colors[i],
            marker = markers[i],
            markersize = 15,
            markerfacecolor = colors[i],
            label = r'$x^{%s}$' % i # f'{i}'
            )
    
plt.xlim(0,10, 50)
plt.ylim(0,100)

plt.xlabel(r'$x$ Value', fontname = 'Times New Roman', fontsize = 32)
plt.ylabel(f'y Value', fontname = 'Times New Roman', fontsize = 32)

plt.xticks(fontsize = 28)
plt.yticks(fontsize = 28)

plt.title(r'$y = f(x) = x^2$', fontname = "Times New Roman", fontsize = 40, y = 1.05)

plt.grid(True)

plt.legend(loc = 'upper left', fontsize = 40)

plt.tight_layout()

```

### Scatter Plot

```python
import csv
from  itertools import groupby

def read_csv(filename):
    with open(filename, 'r') as f:
        data = list(csv.reader(f))
    return data

flowers = read_csv('../files/iris.csv')
flowers

# sepal (leaves at the base of the flower)
# petals
# species
# sepal length (first value), sepal width (second value), petal length (3rd value), last value is the species
# make a scatter plot that shows the sepal length and sepal width for each species of iris
# So I need to split my data in groupings of sepal lengths and sepal widths for each species so that I can plot them

# set up colors for each grouping
colors = {
    'Iris-setosa': '#2B5B84',
    'Iris-versicolor': 'g',
    'Iris-virginica': 'purple'
}
# set up figuresize
plt.figure(figsize=(16,10))

for species, group in groupby(flowers, lambda x: x[4]):
    flower_group = list(group)
    # x-axis: sepal length aka first value in each sublist
    sepal_length = [float(x[0]) for x in flower_group]
    # y-axis: sepal width aka second value in each sublist
    sepal_width = [float(x[1]) for x in flower_group]
    # should have access to the name of the species, a list of the sepal_lengths and a list of the sepal_widths
    # call to plt.scatter([x-values], [y-values], <customization>) to add this set of points to our scatterplot
    plt.scatter(sepal_length, sepal_width, s=40, c=colors[species], label=species)

plt.title('Iris Dataset Sepal Length vs. Width', fontsize=16, y=1.02)
plt.xlabel('Sepal Length (cm)', fontsize=10)
plt.ylabel('Sepal Width (cm)', fontsize=10)
plt.legend(loc='best').get_frame().set_color('chocolate')

plt.show()
```

### Box Plot

```python
# What do box plots tell us? https://miro.medium.com/max/1838/1*2c21SkzJMf3frPXPAR_gZA.png
from itertools import groupby
# visualize distribution of a single variable present in multiple groups
# plot out the distributions of petal lengths for each of our three iris species
# petal lengths are contained in index #2 aka the third value of each flower's data

# list containing a list of petal lengths for each species:
# [[iris-setosa lengths], [iris-virginica lengths], [iris-vesicolor lengths]]
petal_lengths = [[float(x[2]) for x in group] for species, group in groupby(flowers, lambda x: x[4])]
#print(petal_lengths)

# petal_length has no species name attached ->
    # we have to keep track of how petal_lengths was made and therefore which species is which list
    # 1. setosa 2. versicolor 3. virginica

# let's create our boxplot
plt.figure(figsize=(16,10))
# plt.boxplot(<list of lists of values>)
plt.boxplot(petal_lengths)
plt.title('Petal Length of Iris Varietals', fontsize=16, y=1.02)
plt.axis([0, 4, 0, 8]) # specify [xmin, xmax, ymin, ymax]
plt.xticks([1, 2, 3], ['Setosa', 'Versicolor', 'Virginica'])
plt.xlabel('Iris Varietal', fontsize=12)
plt.ylabel('Petal Length (cm)', fontsize=12)

plt.show()
```

### Histograms <br>
<p>While line plots are great for visualizing trends in time series data, histograms are the king of visualizing distributions. Often, the distribution of a variable is what you're interested in, and a visualization provides a lot more information than a group of summary statistics.</p>

```python
fig, ax = plt.subplots(figsize=(16, 9)) #Subplots makes a call in the background to create a plot area

# generate random data - 100 random numbers between 0 and 10
data = np.random.uniform(0,10,100)

# plt.hist(<data>, bins=<# of groups>)
# bins is the number of sections we break our distribution into
# rwidth controls the width of the bar for each bin
plt.hist(data, bins=10, label='Random Number Distribution', color='b', rwidth=0.9)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel('Value', fontsize=20, fontname='Comic Sans MS')
plt.ylabel('Frequency', fontsize=20, fontname='Comic Sans MS')
plt.legend(loc='lower left', fontsize=16)
plt.title('Worst Font Ever Created', fontname='Comic Sans MS', fontsize=24)
plt.show()
```

<p>Let's take a look at a mystery/randomly generated DataFrame, and plot the results</p>

```python
import pandas as pd
import numpy as np

def make_mystery_dataframe():
    customer_id = np.random.randint(1000,1200,10000)
    
    states = {1: 'California', 0: 'Colorado'}
    statelist = []
    agelist = []
    
    for i in range(10000):
        ind = np.random.randint(0,2)
        if ind == 0:
            statelist.append(states[ind])
            agelist.append(np.random.normal(loc = 30, scale = 10))
        elif ind == 1:
            statelist.append(states[ind])
            agelist.append(np.random.normal(loc = 40, scale = 20))
    
    mystery_df = pd.DataFrame.from_dict({
        'customer_id': customer_id,
        'ages':agelist,
        'state': statelist
    })
    
    return mystery_df

df = make_mystery_dataframe()
df.head(20)

coldf = df[df['state'] == 'Colorado']
calidf = df[df['state'] == 'California']
calidf.head()
```

```python
# one plot window containing two separate sublots -> one for cali histogram one for colorado histogram
fig, (col, cali) = plt.subplots(1, 2, figsize=(20, 10))
# here, col is one subplot and cali is the other

# plot for california
# create a histogram of the ages
cali.hist(calidf['ages'], bins=list(range(-25, 126, 10)), label='Age Distribution in California', rwidth=0.8)
# bins can accept a single value and then the size and boundaries of the bins are set automatically
    # or, we can pass a list of boundaries for the bins
cali.set_xlabel('Customer Age', fontname='Times New Roman', fontsize=18)
cali.set_ylabel('Frequency', fontname='Times New Roman', fontsize=18)
cali.set_ylim([0, 2000])
cali.tick_params(axis='both', labelsize=16) # customize the size and appearance of axis ticks
cali.legend(loc='upper right', fontsize=10)

# plot for colorado
col.hist(coldf['ages'], bins=list(range(-25, 126, 10)), label='Age Distribution in Colorado', rwidth=0.8)
col.set_xlabel('Customer Age', fontname='Times New Roman', fontsize=18)
col.set_ylabel('Frequency', fontname='Times New Roman', fontsize=18)
col.set_ylim([0, 2000])
col.tick_params(axis='both', labelsize=16)
col.legend(loc='upper right', fontsize=10)

# overall customization/show
plt.title('Customer Age Distribution by State', fontname='Times New Roman', fontsize=18, y=1.01, x=-.15)
# if you want to save and use elsewhere - save before you show! showing will remove this as the 'active plot'
plt.savefig('../files/foxes84agehistogram.jpg')
plt.show()
```
