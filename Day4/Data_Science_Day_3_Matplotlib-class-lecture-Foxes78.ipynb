### Data Visualization <br>
<p>Data visualization is one of the most powerful tools an analyst has for two main reasons. Firstly, it is unrivalled in it's ability to guide the analysts hand in determining 'what to look at next'. Often a visual is revealing of patterns in the data that are not easily discernable by just looking at dataframes. Secondly, they are an analysts greatest communication tool. Very often professional analysts need to present their results to groups of people responsible for taking action based on what the data says. Visuals can tell your story much better then raw numbers, so let's talk a little bit about how to make effective visuals that make a point.</p>

### Different Types of Charts <br>
<ul>
    <li><b>Line Chart:</b> Exploring data over time</li>
    <li><b>Bar Chart:</b> Comparing categories of data and tracks changes over time</li>
    <li><b>Pie Chart:</b> Explore parts of a whole, i.e. fractions</li>
    <li><b>Scatter Plot:</b> Similar to line charts, track correlations between two categories</li>
    <li><b>Histogram:</b> Unrelated from bar charts, show distribution of variables</li>
    <li><b>Candlestick Chart:</b> Used a lot in financial sector, i.e. can compare a stock over a period of time</li>
    <li><b>Box Chart:</b> Looks identical to candlestick charts, compares minimum, 1st, medium, 3rd quartiles, and max values</li>
</ul>

<p>Depending on what you need to accomplish in conceptualizing your data, you will go with a specific type of chart to portray your data.</p>

### Customization <br>
<p>There's many customizable options when using Matplotlib, such as; line types, colors, labels, figure size, etc. We'll go over a few of them here:</p>

##### Axis Labels & Line Marks

```python
plt.plot([1, 2, 3, 4, 5], [23, 45, 67, 78, 89], '*-y') # format string of '[marker][line][color]'
plt.plot([3, 4, 5, 6, 7, 8], [22, 33, 90, 105, 110, 120], 'x:b')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

##### Legend & Line Labels

```python
plt.plot([1600, 1700, 1800, 1900, 2000], [1, 1.4, 1.6, 1.8, 2], 'o-r', label='Fox')
plt.plot([1600, 1700, 1800, 1900, 2000], [2, 2, 3, 3.1, 2.4], '^:g', label='Chicken')
plt.ylabel('Population (Millions)')
plt.xlabel('Year')
plt.legend() # adds a legend for our line labels
plt.show()
```

### Subplots <br>
<p>Using the subplots method, you're able to create multiple plots within a single cell.</p>

```python
# setting up subplots:
# plt.subplot(rows, columns, position)

# first plot
p1 = plt.subplot(2, 1, 1)
plt.plot([1600, 1700, 1800, 1900, 2000], [1, 1.4, 1.6, 1.8, 2], 'o-r', label='Fox')
plt.ylabel('Population (Millions)')
plt.xlabel('Year')
p1.set_ylim([0, 4])
plt.legend()

# second plot
p2 = plt.subplot(2, 1, 2)
plt.plot([1600, 1700, 1800, 1900, 2000], [2, 2, 3, 3.1, 2.4], '^:g', label='Chicken')
plt.ylabel('Population (Millions)')
plt.xlabel('Year')
p2.set_ylim([0, 4])
plt.legend()

# show both plots
plt.show()
```

### In-Class Exercise #1 - Create a Graph that Depicts y = f(x)

```python
import numpy as np
# use numpy random number generation to create your data
# generate some data as your y values
# come up with a transformation for the x values

# use matplotlib to plot f(x)
x = list(range(10))
y = [i**2 for i in x]
plt.plot(x, y, label='y=x^2')
plt.ylabel('Y')
plt.xlabel('X')
plt.legend()
plt.show()
```

### Presentation Worthiness
<p>Before we go into how to use this as a tool for data exploration, let's talk quickly about presentation-worthiness. In the next cell we're going to introduce customization of lines, colors, ticks, titles, fonts, and grids.</p>

```python
# define some NDarrays to use with the below customization
x_axis = np.linspace(-10,10,num=20)
y_axis = x_axis**2

fig, ax = plt.subplots(figsize = (20, 10))

plt.plot(x_axis ** 2, y_axis ** 4,
        lw = 5, # Line Width
        ls = '--', # Line Style
        color = 'r', # Line Color
        marker = "*", # Pointer Marker
        markersize = 15, # Marker size
        markerfacecolor = 'b', # Marker Color
        label = r'$y=x^2$'
        )

plt.xlabel(r'$x$ Value', fontname = 'Times New Roman', fontsize = 32)
plt.ylabel(r'$y$ Value', fontname = 'Times New Roman', fontsize = 32)

plt.xticks(fontsize = 28)
plt.yticks(fontsize = 28)

plt.title('Quadratic Equation Visual', fontname = "Times New Roman", fontsize = 40, y = 1.05)

plt.grid(True)

plt.legend(loc = 'upper left', fontsize = 40)

plt.tight_layout()
```

<p>It's sometimes nice to obsess a little bit over how your plots look. It's good practice paying attention to detail and it has huge payoff when you end up presenting your findings to other analysts and business parteners. Here's a more complex example of using the matplotlib api:</p>

```python
markers = ['o', 'v', '^', '*', 's']
styles = ['-', '--', '-.', ':', '-']
colors = ['#588C7E', '#F2E394', '#F2AE72', '#D96459', '#8C4646']
```

```python
x = np.linspace(0, 10, 50)

fig = plt.subplots(figsize = (30, 15))

for i in range(5):
    plt.plot(x, x ** i,
            lw = 5,
            ls = styles[i],
            color = colors[i],
            marker = markers[i],
            markersize = 15,
            markerfacecolor = colors[i],
            label = r'$x^{%s}$' % i # f'{i}'
            )
    
plt.xlim(0,10, 50)
plt.ylim(0,100)

plt.xlabel(r'$x$ Value', fontname = 'Times New Roman', fontsize = 32)
plt.ylabel(f'$y$ Value', fontname = 'Times New Roman', fontsize = 32)

plt.xticks(fontsize = 28)
plt.yticks(fontsize = 28)

plt.title(r'$y = f(x)$', fontname = "Times New Roman", fontsize = 40, y = 1.05)

plt.grid(True)

plt.legend(loc='upper left', fontsize=40)

plt.show()

```

### Scatter Plot

```python
import csv
from  itertools import groupby
# useful for showing distributions within groups of data
def read_csv(filename):
    with open(filename, 'r') as f:
        data=list(csv.reader(f))
    return data

flowers = read_csv('../files/iris.csv')

# I want 3 of these values - sepal length (first value), sepal width (second value), species (last value)

# before I organize and plot my data, I need to do certain customizations before the main plotting
    # main plotting includes plt.plot(), plt.scatter(), etc.
    # customizations that need to come before main plotting include:
        # figure size
        # subplots/subplot orientation
        # certain axis customization
# set up my colors
colors = {
    'Iris-setosa': '#2B5B84',
    'Iris-versicolor': 'g',
    'Iris-virginica': 'purple'
}
# set up my figuresize
plt.figure(figsize=(16,10))

# I need to organize the data, then do a plt.scatter() for each iris type
# I'm going to use itertools.groupby()
for species, group in groupby(flowers, lambda x: x[4]):
    flower_group = list(group)
    # x-axis: sepal length aka the first value in each sublist -> list of x values aka a list of the first indexes from each flower
    sepal_length = [float(x[0]) for x in flower_group]
    # y-axis: sepal width aka the second value in each sublist -> list of second indexes from each flower
    sepal_width = [float(x[1]) for x in flower_group]
    # plt.scatter()
    plt.scatter(sepal_length, sepal_width, s=40, c=colors[species], label=species)

plt.title('Iris Dataset Sepal Length vs. Width', fontsize=16, y=1.02)
plt.xlabel('Sepal Length (cm)', fontsize=10)
plt.ylabel('Sepal Width (cm)', fontsize=10)
plt.legend(loc='best').get_frame().set_color('chocolate')

plt.show()
```

### Box Plot

```python
# what do boxplots show us? https://miro.medium.com/max/1838/1*2c21SkzJMf3frPXPAR_gZA.png

from itertools import groupby
# visualizing distributions of a single variable among multiple groups
# plot out the distributions of petal lengths for each of our three iris species
# petal lengths are in index #2 aka the third index of each flower's data

# petal_length is a list containing a list of petal lengths for each species
petal_length = []
for species, group in groupby(flowers, lambda x: x[4]):
    #print(species)
    petal_length.append([float(x[2]) for x in group])

# petal_length has no species names attached -> we can keep track of which is which based on indexes from our groupby

# let's create our boxplot
plt.figure(figsize=(16,10))
# plt.boxplot()
plt.boxplot(petal_length)
plt.title('Petal Length of Different Iris Varietals', fontsize=18, y=1.02)
plt.axis([0, 4, 0, 8]) # xmin, xmax, ymin, ymax
plt.xticks([1, 2, 3], ['Setosa', 'Versicolor', 'Virginica'])
plt.xlabel('Iris Varietal', fontsize=12)
plt.ylabel('Petal Length(cm)', fontsize=12)

plt.show()
```

### Histograms <br>
<p>While line plots are great for visualizing trends in time series data, histograms are the king of visualizing distributions. Often, the distribution of a variable is what you're interested in, and a visualization provides a lot more information than a group of summary statistics.</p>

```python
fig, ax = plt.subplots(figsize=(16, 9)) #Subplots makes a call in the background to create a plot area

# generate random data
data = np.random.uniform(0,10,100)

# plt.hist()
plt.hist(data, bins=10, label='Random Number Distribution', color='b', rwidth=0.9)
# bins are the number of sections we break our distribution into
# rwidth lets us set the percentage of the bin that the bar actually takes up
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel('Value', fontsize=20, fontname='Comic Sans MS')
plt.ylabel('Frequency', fontsize=20, fontname='Comic Sans MS')
plt.legend(loc='lower left', fontsize=16)
plt.title('Worst Font Ever Created', fontname='Comic Sans MS', fontsize=20)
plt.show()
```

<p>Let's take a look at a mystery/randomly generated DataFrame, and plot the results</p>

```python
import pandas as pd
import numpy as np

def make_mystery_dataframe():
    customer_id = np.random.randint(1000,1200,10000)
    
    states = {1: 'California', 0: 'Colorado'}
    statelist = []
    agelist = []
    
    for i in range(10000):
        ind = np.random.randint(0,2)
        if ind == 0:
            statelist.append(states[ind])
            agelist.append(np.random.normal(loc = 30, scale = 10))
        elif ind == 1:
            statelist.append(states[ind])
            agelist.append(np.random.normal(loc = 40, scale = 20))
    
    mystery_df = pd.DataFrame.from_dict({
        'customer_id': customer_id,
        'ages':agelist,
        'state': statelist
    })
    
    return mystery_df

df = make_mystery_dataframe()
df.head(20)

# i need a list of the ages from each state
# 1. filter the df for a single state
# 2. select just the ages
cali = df[df['state']=='California']['ages']
col =  df[df['state']=='Colorado']['ages']
```

```python
# goal is two side by side histograms showing customer age distribution in each state
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10)) # how many subplots and how are they oriented

# plot for california
# for bins -> a list of the borders of the bins
ax1.hist(cali, bins=list(range(-40, 131, 10)), label='Age Distribution in California', rwidth=0.8)
ax1.set_xlabel('Customer Age', fontname='Times New Roman', fontsize=18)
ax1.set_ylabel('Frequency', fontname='Times New Roman', fontsize=18)
ax1.set_ylim([0, 1750])
ax1.tick_params(axis='both', labelsize=16)
ax1.legend(loc='upper right', fontsize=10)

# plot for colorado
ax2.hist(col, bins=list(range(-40, 131, 10)), label='Age Distribution in Colorado', rwidth=0.8)
ax2.set_xlabel('Customer Age', fontname='Times New Roman', fontsize=18)
ax2.set_ylabel('Frequency', fontname='Times New Roman', fontsize=18)
ax2.set_ylim([0, 1750])
ax2.tick_params(axis='both', labelsize=16)
ax2.legend(loc='upper right', fontsize=10)

# overall customization/show
plt.title('Customer Age Distribution by State', fontname='Times New Roman', fontsize=18, y=1.02, x=-.17)
# if saving, save before showing as showing will remove this as the active plot
plt.savefig('../files/customeragehistogram.jpg')
plt.show()
```
